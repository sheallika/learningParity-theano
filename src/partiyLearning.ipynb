## Learning Parity Function

```python
import theano
from layers import test_mlp_parity,test_rnn_parity,test_lstm_parity
```

### Learning Parity Function
#### Bullet 1: Shallow MLP

```python
n_bit=[8,12]
n_layers=[800,1200]
for i,k in zip(n_bit,n_layers):
    
    test_mlp_parity(n_bit=i, n_hidden=k, n_hiddenLayers=1, n_epochs=1000, batch_size=50, verbose= True)
    
```

#### Bullet 2: Deep MLP

```python
n_bit=[8,12]
n_layers=[[100,50,50],[100,100,50]]
for i,k in zip(n_bit,n_layers):
    
    test_mlp_parity(n_bit=i, n_hidden=k, n_hiddenLayers=3, n_epochs=1000, batch_size=50, verbose= True)
    
```

#### Bullet 3: RNN

```python
n_bit=[8,12]
nhidden=[3,7]
for i,j in zip(n_bit,nhidden):
    test_rnn_parity(n_bit=i,nhidden=j,nepochs=100)
```

#### Bullet 4: LSTM

```python
n_bit=[8,12]
nhidden=[1,1]
for i,j in zip(n_bit,nhidden):
    test_lstm_parity(n_bit=i,nhidden=j,nepochs=60)
```
